```python
import json
file = "/media/chris/UNTITLED/Beauty_and_Personal_Care.jsonl" # Path to input file
output_path = "filtered.json" # Path to output file
ids = ['B01K1HPA60', 'B0BJMV1QTR','B0BJLPRXZ6','B0BLGN9N39'] # List of items to filter
print_mark = 1000000 #At which number of lines to print output
```

```python
full_data = []
with open(file, 'r') as fp:    
        for x,line in enumerate(fp):
            data = json.loads(line.strip())
            if data["parent_asin"] in ids:
                full_data.append(data)

            if x%print_mark==0:
                print(f"Finished {x} lines")

            
with open(output_path, "w") as outfile:
    json.dump(full_data, outfile)
```
